```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# 1. Crea il "cervello" del modello
model = LinearRegression()

# 2. ADDESTRA IL MODELLO (Fit)
# Qui il modello "studia" le relazioni tra i numeri scalati e il prezzo
print("Addestramento in corso...")
model.fit(X_train_final, y_train)

# 3. FAI LE PREVISIONI (Predict)
# Ora lo mettiamo alla prova sui dati che NON ha mai visto (Test set)
y_pred = model.predict(X_test_final)

# 4. VEDIAMO I RISULTATI (Score)
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("\n--- RISULTATI ---")
print(f"R2 Score (Precisione da 0 a 1): {r2:.4f}")
print(f"RMSE (Errore medio logaritmico): {rmse:.4f}")
```

```python
train_score = model.score(X_train_final, y_train)
test_score = r2  # Quello che hai già calcolato

print(f"Punteggio Train: {train_score:.4f}")
print(f"Punteggio Test:  {test_score:.4f}")

if train_score - test_score < 0.05:
    print("VERDETTO: Modello solidissimo! Niente Overfitting.")
else:
    print("VERDETTO: Occhio, c'è un po' di Overfitting.")
```
